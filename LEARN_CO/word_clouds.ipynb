```python
#Reading in a text file
filename = 'beach_boys_baran.txt'
text = ""
with open(filename) as f: #Opens the file with an alias.
    for line in f: #Iterate line by line through the file
        text+=line #Add the line to the text string
print(len(text)) #Print how many characters (including spaces, enters, etc.) are in the string
print(text[:50]) #Print the first 50 characters of the string
```

```python
cloud = wordcloud.WordCloud().generate(text)
plt.imshow(cloud)
```
